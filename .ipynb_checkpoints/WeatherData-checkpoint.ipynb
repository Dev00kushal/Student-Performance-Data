```python
import pandas as pd 
import numpy as np
import matplotlib as plt
import seaborn as sns

df = pd.read_csv('./data/weatherAUS.csv')

df.head()
```

```python
print(df.info())
```

# Cleaning Data

## Remove Rows

```python
df.dropna(inplace = True)

df.head()
```

```python
# Checking for duplicate data and removing them if any
print(df.duplicated())
df.drop_duplicates(inplace = True)
```

## Separating Numeric column and Categorical column from df

### Numeric Column

```python
# Separate numeric columns
numeric_df = df.select_dtypes(include=[np.number])
numeric_df.head()
```

### Categorical Column

```python
# Separate categorical columns
categorical_df = df.select_dtypes(exclude=[np.number])
categorical_df.head()
```

## Replace Empty Values with Mean

```python
# Replace empty numeric value with mean
numeric_df.fillna(numeric_df.mean(), inplace = True)
numeric_df.head()
```

```python
# Replace empty categorical value with mode
categorical_df.fillna(categorical_df.mode(), inplace = True)
categorical_df.head()
```
